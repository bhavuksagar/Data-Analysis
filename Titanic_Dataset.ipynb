```python
# checking null values in data
data.isnull()
```

```python
#sum of NaN values

data.isnull().sum()
```

```python
#filling the null values and dropping the useless columns
data.Age.fillna(data.Age.mean(),inplace=True)
```

```python
data.isnull().sum()
```

```python
data.drop(["Cabin","Ticket","Name"],axis=1,inplace=True)
```

```python
data.head()
```

```python
print(data['Embarked'].value_counts())
```

```python
# filling nan value of Embarked column
data.Embarked.fillna("S",inplace=True)
```

```python
data.isnull().sum()
```

```python
#handling the strings
def change(s):                #changing the string into number
    if s=="female":
        return 0
    else:
        return 1
data["sex"]=data.Sex.apply(change) 
```

```python
data.head()
```

```python
del data["Sex"]
```

```python
data.head()
```

```python
def change(s):                #changing the string into number
    if s=="S":
        return 1
    else:
        return 0
data["Embarked_S"]=data.Embarked.apply(change) 

def change(s):                #changing the string into number
    if s=="Q":
        return 1
    else:
        return 0
data["Embarked_Q"]=data.Embarked.apply(change) 

def change(s):                #changing the string into number
    if s=="C":
        return 1
    else:
        return 0
data["Embarked_C"]=data.Embarked.apply(change) 
```
