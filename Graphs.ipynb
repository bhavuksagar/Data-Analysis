# Customising the Graph

```python
#function used- grid(), xlabel(),ylable(),legend(),title(),text(),axis()

x=np.arange(0,5,0.01)     
y=x**3
y2=x**2
pl.plot(x,y,color="black",marker="o",linewidth=2, label="X**3") #we can give color and marker by color and marker function and 
                                                                #give the name to the line by label and also adjust the size of 
                                                                #the line
pl.plot(x,y2,color="red",marker="o",linewidth=2,label="X**2")

pl.grid()                                                       #to add grids on the graph

pl.text(3,60,"Point",fontsize=20)                               #to add text on the particular point.parameter contain the 
                                                                #coordinates and the text

pl.xlabel("X")                                                  #use for give the name to the x and y 
pl.ylabel("Y")

pl.title("X^3 and X^2 Graph")                                   #give the title to the graph

pl.axis([0,6,0,140])                                            #make ur own axis.it contain 4 parameter first two for x-axis 
                                                                #and other two for y-axis
pl.legend()                                                     #give the detail of line on the graph
pl.show()
```

# pie Graph

```python
z=[3,4,7,8]
lab=["A","B","C","D"]  #give the names to the blocks
explod=[.2,0,0,0]    #take out the one piece from the whole circle
color=["blue","red","purple","green"]  #change the colors
pl.pie(z,colors=color,explode=explod,labels=lab,counterclock=False,autopct="%.2f%%",startangle=100) 
pl.axis("equal")   #correct the size or create the proper circle
pl.title("Pie Graph")  #give the tile name
pl.show()

#counterclock=rotation in clockwise
#autopct="%.2f%%"-> use for give the percentage to each block
#startangle= to change the starting angle
```
